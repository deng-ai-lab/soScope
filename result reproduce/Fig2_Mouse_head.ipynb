```python
def norm(data):
    _range = np.max(data) - np.min(data)
    return (data - np.min(data)) / _range
 
 
def standardization(data):
    mu = np.mean(data, axis=0)
    sigma = np.std(data, axis=0)
    return (data - mu) / sigma

pearson_dic = {}
mse_dic = {}
norm_mse_dic = {}

for key in method_names[1:]:
    pearson_dic[key] = []
    mse_dic[key] = []
    norm_mse_dic[key] = []
    
top_gene = gene_name
for i in range(len(top_gene)):
    gene = top_gene[i]
    exp_gt = data_dic['reordered_count'][gene]
    for key in method_names[1:]:
        exp_ = data_dic[key][gene]
        # pearson
        pearson_, _ = pearsonr(exp_gt, exp_)
        # mse
        mse_ = mean_squared_error(exp_gt, exp_)
        norm_mse_ = mean_squared_error(norm(exp_gt), norm(exp_))
        pearson_dic[key].append(pearson_)
        mse_dic[key].append(mse_)
        norm_mse_dic[key].append(norm_mse_)
```

```python
norm_mse_df = pd.DataFrame([(method, mse) for method, mses in norm_mse_dic.items() for mse in mses], 
                       columns=['Methods', 'MSE'])

plt.figure(figsize=(10, 6))
sns.boxplot(x='Methods', y='MSE', data=norm_mse_df, order=order)
plt.ylim(0, 0.3)
plt.tight_layout()
plt.savefig(f'{project_name}/mse.pdf', pad_inches=0)
plt.show()
```

```python
def cluster_sort_df(df):
    corr = df.corr()
    distance_matrix = 1 - corr.abs()
    linkage_matrix = linkage(squareform(distance_matrix), method='ward')
    leaves = leaves_list(linkage_matrix)
    sorted_df = df.iloc[:, leaves]
    return sorted_df
```

```python
def get_corr(data_dir, genes):
    cor_high_mat_list = []
    for key in data_dic.keys():
        df = data_dic[key][genes]
        data = df.values.astype(np.float32)
        corr_matrix = np.corrcoef(data, rowvar=False)
        cor_high_mat_list.append(corr_matrix)
    return cor_high_mat_list
```

```python
def calculate_metrics(matrix1, matrix2):
    mse = np.mean((matrix1 - matrix2) ** 2)
    pearson_corr, _ = pearsonr(matrix1.flatten(), matrix2.flatten())
    return mse, pearson_corr

# high
high_bio = top_genes[:60]
cor_high_mat_list = get_corr(data_dic, high_bio)

matrix1 = cor_high_mat_list[0]
methods = cor_high_mat_list[1:]
metrics = [calculate_metrics(matrix1, matrix) for matrix in methods]

mses_high, pearsons_high = zip(*metrics)

# mid
mid_bio = top_genes[60:120]
cor_mid_mat_list = get_corr(data_dic, mid_bio)

matrix1 = cor_mid_mat_list[0]
methods = cor_mid_mat_list[1:]
metrics = [calculate_metrics(matrix1, matrix) for matrix in methods]

mses_mid, pearsons_mid = zip(*metrics)

# low
low_bio = top_genes[120:194]
cor_low_mat_list = get_corr(data_dic, low_bio)
matrix1 = cor_low_mat_list[0]
methods = cor_low_mat_list[1:]
metrics = [calculate_metrics(matrix1, matrix) for matrix in methods]

mses_low, pearsons_low = zip(*metrics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.spatial.distance import squareform

top_genes = gene_info_sorted['Gene'].tolist()[:194]

def plot_corr_matrix(df, title, ax):
    data = df.values.astype(np.float32)
    corr_matrix = np.corrcoef(data, rowvar=False)
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', square=True, xticklabels=False, yticklabels=False, vmin=-1, vmax=1, ax=ax)
    ax.set_title(title)

fig, axes = plt.subplots(3, len(data_dic), figsize=(24, 12))

for j, gene_range in enumerate([top_genes[:60], top_genes[60:120], top_genes[120:194]]):
    data_list = [data_dic[k][gene_range] for k in ['reordered_count']+order]
    key_list = list(data_dic.keys())
    sorted_df1 = cluster_sort_df(data_list[0])
    sorted_columns = sorted_df1.columns
    sorted_data_list = [df[sorted_columns] for df in data_list]

    for i, df in enumerate(sorted_data_list):
        ax = axes[j, i]
        plot_corr_matrix(df, f"Group: {j+1}, Method: {(['reordered_count']+order)[i]}", ax)

plt.tight_layout()

plt.savefig(f'{project_name}/correlation_heatmaps.pdf', pad_inches=0)

plt.show()
```
